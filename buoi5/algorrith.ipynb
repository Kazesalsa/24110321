Code BFS

```python
import random
from collections import deque

def newnode(state, parent=None, move=None):
    return {"state": state, "parent": parent, "move": move}

def get_solution(node):
    path = []
    while node:
        path.append(node)
        node = node["parent"]
    return path[::-1]

def bfs(initial, goal):
    node = newnode(initial)
    if initial == goal:
        return get_solution(node)

    frontier = deque([node])
    frontier_states = {initial}
    explored = set()

    while frontier:
        node = frontier.popleft()
        state = node["state"]
        frontier_states.remove(state)
        explored.add(state)

        row, col = findd(state)
        moves = []
        if row > 0: moves.append('U')
        if row < 2: moves.append('D')
        if col > 0: moves.append('L')
        if col < 2: moves.append('R')
        random.shuffle(moves)

        for move in moves:
            next_state = move_blank(state, move, row, col)
            if next_state not in explored and next_state not in frontier_states:
                child = newnode(next_state, node, move)
                if next_state == goal:
                    return get_solution(child)
                frontier.append(child)
                frontier_states.add(next_state)

    return None

def findd(state):
    for r in range(3):
        for c in range(3):
            if state[r][c] == 0:
                return r, c

def move_blank(state, move, r, c):
    a = [list(row) for row in state]
    if move == 'U':
        a[r][c], a[r-1][c] = a[r-1][c], a[r][c]
    elif move == 'D':
        a[r][c], a[r+1][c] = a[r+1][c], a[r][c]
    elif move == 'L':
        a[r][c], a[r][c-1] = a[r][c-1], a[r][c]
    elif move == 'R':
        a[r][c], a[r][c+1] = a[r][c+1], a[r][c]
    return tuple(tuple(row) for row in a)

start = (
    (2, 8, 3),
    (1, 6, 4),
    (7, 0, 5)
)

goal = (
    (1, 2, 3),
    (8, 0, 4),
    (7, 6, 5)
)

path = bfs(start, goal)

if path:
    print(f"tong so buoc: {len(path) - 1}")
    for node in path:
        if node["move"]:
            print(node['move'])
        for row in node["state"]:
            print(" ".join(map(str, row)))
```

```python
import random

def newnode(state, parent=None, move=None):
    return {"state": state, "parent": parent, "move": move}

def get_solution(node):
    path = []
    while node:
        path.append(node)
        node = node["parent"]
    return path[::-1]

def dfs(initial, goal):
    node = newnode(initial)
    frontier = [node]
    explored = set()

    while frontier:
        node = frontier.pop()
        state = node["state"]

        if state == goal:
            return get_solution(node)

        if state not in explored:
            explored.add(state)

            row, col = findd(state)
            moves = []
            if row > 0: moves.append('U')
            if row < 2: moves.append('D')
            if col > 0: moves.append('L')
            if col < 2: moves.append('R')
            random.shuffle(moves)

            for move in moves:
                next_state = move_blank(state, move, row, col)
                child = newnode(next_state, node, move)
                frontier.append(child)

    return None

def findd(state):
    for r in range(3):
        for c in range(3):
            if state[r][c] == 0:
                return r, c

def move_blank(state, move, r, c):
    a = [list(row) for row in state]
    if move == 'U':
        a[r][c], a[r-1][c] = a[r-1][c], a[r][c]
    elif move == 'D':
        a[r][c], a[r+1][c] = a[r+1][c], a[r][c]
    elif move == 'L':
        a[r][c], a[r][c-1] = a[r][c-1], a[r][c]
    elif move == 'R':
        a[r][c], a[r][c+1] = a[r][c+1], a[r][c]
    return tuple(tuple(row) for row in a)

start = (
    (2, 8, 3),
    (1, 6, 4),
    (7, 0, 5)
)

goal = (
    (1, 2, 3),
    (8, 0, 4),
    (7, 6, 5)
)

path = dfs(start, goal)

if path:
    print(f"tong so buoc: {len(path) - 1}")
    for node in path:
        if node["move"]:
            print(node['move'])
        for row in node["state"]:
            print(" ".join(map(str, row)))
```
